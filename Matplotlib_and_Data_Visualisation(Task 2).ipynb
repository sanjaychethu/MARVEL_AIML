```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the growth of a plant
time = np.arange(0, 10, 0.1)
growth = np.exp(time)

# Plot the growth
plt.plot(time, growth, color='green')

# Set creative axes limits
plt.xlim(0, 8)  # Setting x-axis limit to 0 to 8
plt.ylim(0, 500)  # Setting y-axis limit to 0 to 500

# Add labels and title
plt.xlabel('Time (days)')
plt.ylabel('Plant Height (cm)')
plt.title('Exponential Growth of a Plant')

# Show plot
plt.grid(True)  # Adding grid for better visualization
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for time and distance traveled by a space probe
time = np.arange(0, 10, 0.5)
distance = np.sin(time) * 100

# Line plot
plt.plot(time, distance, label='Distance', color='red')
plt.title('Distance vs Time')
plt.xlabel('Time (hours)')
plt.ylabel('Distance (millions of km)')
plt.legend()
plt.show()

# Area plot
plt.fill_between(time, distance, color='purple', alpha=0.5)
plt.title('Distance vs Time (Area Plot)')
plt.xlabel('Time (hours)')
plt.ylabel('Distance (millions of km)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load Iris dataset
iris = sns.load_dataset("iris")

# Scatter plot
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=iris['petal_length'], cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Size vs Petal Length (Color Encoded)')
plt.colorbar(label='Petal Length')
plt.show()

# Bubble plot
plt.scatter(iris['sepal_length'], iris['sepal_width'], s=iris['petal_width']*100, alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Size vs Petal Width (Bubble Plot)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for satellite launch success rate
launch_success = np.random.normal(0.8, 0.1, 1000)  # Mean: 0.8, Std. deviation: 0.1

# Histogram
plt.hist(launch_success, bins=20, color='salmon', edgecolor='black')
plt.xlabel('Success Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Satellite Launch Success Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for payload capacities of satellite launches (in kg)
payloads = [np.random.normal(1500, 300, 100) for _ in range(4)]

# Box plot
plt.boxplot(payloads, labels=['GEO', 'LEO', 'MEO', 'HEO'])
plt.xlabel('Orbit')
plt.ylabel('Payload Capacity (kg)')
plt.title('Satellite Payload Capacity by Orbit')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris dataset
iris = sns.load_dataset("iris")

# Violin plot
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

```python
import seaborn as sns

# Load Iris dataset
iris = sns.load_dataset("iris")

# Marginal plot
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='kde')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Joint Distribution of Petal Length and Width')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for contour plot
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Contour plot
plt.contour(X, Y, Z, cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of sin(X) * cos(Y)')
plt.colorbar(label='Z')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Generate random data for demonstration
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Customize labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot with Random Data')

plt.show()
```
